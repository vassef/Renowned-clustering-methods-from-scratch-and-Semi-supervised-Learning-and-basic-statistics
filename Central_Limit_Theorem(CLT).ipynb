```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def pick_sample(dist):
  fig,axes=plt.subplots(2,4,figsize=(20,10),dpi=85)
  lst=[1,10,40,64,80,100,500,1000]
  for i,num in enumerate(lst):
    new_dist=dist.reshape((int(10**6/num),num))
    mean_dist=np.mean(new_dist,axis=1)
    sns.histplot(mean_dist,ax=axes[int(i/4),i%4])
    axes[int(i/4),i%4].set_title(f"number of samples : {num},with mu : {np.round(np.mean(mean_dist),2)},and std:{np.round(np.std(mean_dist),2)}")
  plt.tight_layout()
```

# Exponential disturbution

```python
exp_dist=np.random.exponential(scale=0.5,size=(1,10**6))
```

```python
pick_sample(exp_dist)
```

# Bionomial Disturbution

```python
bio_dist=np.random.binomial(20,0.8,size=(1,10**6))
```

```python
pick_sample(bio_dist)
```
