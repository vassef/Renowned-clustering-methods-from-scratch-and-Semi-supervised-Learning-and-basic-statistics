```python
from random import randint
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import numpy as np
from multiprocessing import Pool, cpu_count

def random_birthdays(number_of_people):
    first_day_of_year = datetime(2022, 1, 1)
    return [first_day_of_year + timedelta(days=randint(0, 365)) for i in range(number_of_people)]
```

```python
random_birthdays(23)
```

```python
def determine_probability(number_of_people, run_amount=10000):
    dups_found = 0
    for i in range(run_amount):
        birthdays = random_birthdays(number_of_people)
        duplicates = set(x for x in birthdays if birthdays.count(x) > 1)
        #print(duplicates)
        if len(duplicates) >= 1:
            dups_found += 1

    return dups_found/run_amount * 100
```

```python
determine_probability(0)
```

```python
def plot_yearly_probabilities(max_people, vertical_markers=(23, 70)):
  plt.figure(figsize=(12,6))
  plt.grid()
  prob=np.zeros(max_people)
  for i,num in enumerate(range(max_people)):
    prob[i]=determine_probability(num)
    if num in vertical_markers:
       plt.axvline(num, color='red',ls='--')
  plt.plot(np.arange(0,max_people),np.sort(prob),marker='o',color='blue')
  plt.xlabel('number of people',size=13)
  plt.ylabel('Probability of having at least two same birthday')
```

```python
#~3 min
plot_yearly_probabilities(100)
```
